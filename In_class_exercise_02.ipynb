(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
# write your answer here
import numpy as np
arr=np.array([1,2,3,4,5])
arr1=np.array([1,1,2,3,4,4])
def dupli(arr):
  temp=0
  for i in range(0,len(arr)):
    for j in range(i+1,len(arr)):
      if arr[i]==arr[j]:
        temp=temp+1
  if temp==0:
    return -1

print("arr",dupli(arr))
print("arr1",dupli(arr1))


    


  




```

(2) Write a Python program to select all the Sundays of a specified year. (4 points)



```python
# write your answer here
import datetime
import calendar
cal= calendar.Calendar()
sundays=[]
x=cal.yeardatescalendar(2020, width=12)
for i in x[0]:
  for j in i:
    for k in j:
      if k.strftime("%A")=="Sunday":
        sundays.append(k)

print(sundays)
```

*   Write a program to read all the txt files and save the sentences in all the files into one csv file with two columns, the first column is sentence id (txt file name+sentence line number), the second column is the sentence text content. (4 points)
*   Remove all the punctuations from the sentences, save the processed sentences into a new column in the same csv file. (4 points)
*   Ask the user to enter a word, return all the sentences that include this word, three kinds of information should be returned: sentence id, sentence text content, the count that user input word appear in the sentence. (4 points)



```python
# write your answer here
import os             
import pandas as pd
import nltk
nltk.download("punkt")
all_files = os.listdir("/content/drive/My Drive/exercise_02_data_collection")   # imagine you're one directory above test dir
col1=[]
col2=[]
col3=[]
count=0
df3=[]
for i in range(0,len(all_files)):
  temp=1
  f=open("/content/drive/My Drive/exercise_02_data_collection"+"/"+all_files[i],mode='r')
  for line in f:
    col1.append(all_files[i]+" "+str(temp))
    col2.append(line)
    df2words = nltk.word_tokenize(line)
    col3.append([w for w  in df2words if w.isalnum()])
    temp=temp+1
  df2=pd.DataFrame(data={'col1':col1,'col2':col2,'col3':col3})
word_to_search=input("enter word to be searched:")
for i in range(len(df2['col3'])):
  if word_to_search in df2['col3'][i]:
    df3.append([df2['col1'][i],df2['col2'][i]])
    count=count+1
print(df2)
print(word_to_search+" appeared in"+" "+str(count)+" sentences")
print(df3)
print("count = "+ str(count))

```
